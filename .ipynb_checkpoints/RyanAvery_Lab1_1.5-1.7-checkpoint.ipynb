```python
import pandas as pd
%matplotlib inline
neuse = pd.read_csv('neuse_metric.txt', delimiter='\t', header=None)
neuse = neuse.rename(columns={0:'time',1:'value'})
neuse.plot(x='time', y='value')
```

## 1.5 Checking for Zeros

```python
def check_zero(df):
    for val in df.value:
        if val == 0:
            print('We got a zero!')
    print('Done checking for zeros')
    return
check_zero(neuse)
```

```python
brf = pd.read_csv('brf_temp-1.6.txt', delimiter='\t', header=None)
brf = brf.rename(columns={0:'time',1:'value'})
brf.plot(x='time', y='value')
```

## 1.6 Filtering Bad Values and finding largest hourly change

```python
brf['difference'] = brf.diff(periods=1).value # the .diff method gets rate of change
for i, diffval in enumerate(brf.difference):
    if abs(diffval) > 6:
        brf.value[i] = None 
        brf.difference[i] = None
    if abs(brf.value[i]) > 40:
        brf.value[i] = None 
```

I removed the clear drops and spikes with the above for loop, setting absolute differences above 6 and absolute temp values above 40 to None (No Data). Below is the filtered plot. The value of 6 was determined after examining the histogram and plots below.

```python
brf.plot(x='time', y='value')
```

And below is a histogram and plot of the difference values. It looks like there still may be some outliers that could be erroneous.

```python
brf.plot(x='time', y='difference', kind='hist', bins=100)
```

```python
brf.plot(x='time', y='difference')
```

We can use a standard deviation filter to remove temp values where the rate of change has a standard deviation greater than 4.

```python
import numpy as np
stddev= np.sqrt(np.sum((brf.difference-brf.difference.mean())**2)/len(brf.difference))
```

```python
brf['stddevs']=brf.difference/stddev
brf.plot(x='time',y='stddevs')
```

```python
for i, std in enumerate(brf.stddevs):
    if abs(std) > 4:
        brf['value'][i] = None 
        brf['difference'][i] = None
brf.plot(x='time', y='difference', kind='hist', bins=100)
```

```python
brf.plot(x='time', y='difference')
```

```python
brf.plot(x='time', y='value')
```

## After filtering, the largest hourly rate of change appears to be 3.62 degrees Celsius.
